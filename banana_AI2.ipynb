```python
from ultralytics import YOLO

# Load a YOLOv8 model (you can use yolov8n.pt, yolov8s.pt, yolov8m.pt, etc.)
model = YOLO("yolov8n.pt")  # 'n' is the smallest, fastest version

# Train the model
model.train(
    data="data.yaml",  # Replace with your correct .yaml path
    epochs=30,
    imgsz=640,
    batch=16
)
```

```python
from ultralytics import YOLO

model = YOLO("runs/detect/train/weights/best.pt")
model.export(format="onnx")         # Export to ONNX
model.export(format="torchscript")  # Export to TorchScript
model.export(format="coreml")       # For iOS (CoreML)
model.export(format="engine")       # TensorRT (NVIDIA devices)
```
